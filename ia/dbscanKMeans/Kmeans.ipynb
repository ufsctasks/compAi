```python
# Importar bibliotecas necessárias
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import plotly.express as px

# Carregar os dados do dataset Galton Families
galton_families = pd.read_csv('/content/sample_data/GaltonFamilies.csv')

# Exibir as primeiras linhas do dataset
print(galton_families.head())

# Separar os dados em variáveis de entrada (X)
X = galton_families[['father', 'mother', 'childHeight']].values

# Aplicar KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans_labels = kmeans.fit_predict(X)

# Visualizar os clusters formados pelo KMeans
grafico_kmeans = px.scatter_3d(x=X[:, 0], y=X[:, 1], z=X[:, 2], color=kmeans_labels, labels={'x': 'Father Height', 'y': 'Mother Height', 'z': 'Child Height'})
grafico_kmeans.show()

# Analisar os resultados do KMeans
print("Labels do KMeans:", np.unique(kmeans_labels, return_counts=True))
```
